```python
breakpoint = result[result['Model'] =='Randomforest classifier - Tuned(gridsearch)'].index[0]+1
firstpart=result.iloc[:breakpoint]
last_5rows=result.iloc[-5:]
middle=result.iloc[breakpoint:-5]
new_order_df = pd.concat([firstpart, last_5rows, middle], ignore_index=True)
new_order_df.reset_index(drop=True, inplace=True)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Round the error metric columns to two decimal places
metrics = ["Recall of Fraud", "F1 Score", "ROC Score", "Accuracy"]
new_order_df[metrics] = new_order_df[metrics].round(2)

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(14, 10))  # Increased figure size for better readability
ax.axis('tight')
ax.axis('off')

# Create the table with formatted data
table = ax.table(cellText=new_order_df.values, colLabels=new_order_df.columns, cellLoc='center', loc='center')

# Adjust font size, row height, and column width
table.auto_set_font_size(False)
table.set_fontsize(10)  # Larger font size for readability
table.scale(1.5, 1.5)  # Scale table for better fit

# Set specific column width for "Model" and row height for all rows
table.auto_set_column_width([0]) 
table.auto_set_column_width([2])
table.auto_set_column_width([3])
table.auto_set_column_width([4])# Adjust the width of the "Model" column
table.auto_set_column_width([5])
for i in range(len(new_order_df) + 1):  # +1 for the header row
    table[i, 0].set_width(0.3)  # Increase width of "Model" column
    for j in range(len(new_order_df.columns)):
        table[i, j].set_height(0.08)  # Set row height

# Save the table as a PDF
plt.savefig("creditcard_result_HW.pdf", bbox_inches='tight')
```
